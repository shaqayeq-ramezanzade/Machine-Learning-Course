```python
#too formule Scurve bayaed b1 va b2 hesab beshe k shib khat va jabejayee she
def sigmoid(x, b1, b2):
    y = 1 / (1 + np.exp(-b1 * (x - b2)))
    return y
```

```python
from scipy.optimize import curve_fit
popt,pcov= curve_fit(sigmoid, xdata, ydata)
#print final parameters 
print("beta_1=%f, beta_2=%f"%(popt[0],popt[1])) #popt=optimozed parameter 0 va 1
```

# Hirerchial Lab1!

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from scipy import ndimage
from sklearn import manifold, datasets
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
%matplotlib inline
```

```python
#make a sample
X1,y1= make_blobs(n_samples=50, centers=[[4,4],[-2,-1], [1,1], [10,4]],cluster_std=0.9)
```

```python
print(X1)
```

```python
#plot X1
plt.scatter(X1[:,0],X1[:,1],marker='.')
```

```python
#make Agglomerative clustering by sikitlearn
aggl=AgglomerativeClustering(n_clusters=4, linkage='average')
```

```python
#fit
aggl.fit(X1,y1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure of size 6 inches by 4 inches.
plt.figure(figsize=(6, 4))

# Scale the data points in X1 to a range of [0, 1].
X1 = (X1 - np.min(X1, axis=0)) / (np.max(X1, axis=0) - np.min(X1, axis=0))

# Display each data point with its cluster value and color-coded using a colormap.
for i in range(X1.shape[0]):
    plt.text(X1[i, 0], X1[i, 1], str(y1[i]),
             color=plt.cm.nipy_spectral(aggl.labels_[i] / 10.),
             fontdict={'weight': 'bold', 'size': 9})



# Display the plot of the original data before clustering.
plt.scatter(X1[:, 0], X1[:, 1], marker='.')

# Display the plot.
plt.show()
```

```python
#lest make dendrogram
dist_matrix=distance_matrix(X1,X1)
dist_matrix #avalish 0 chon mishe fasele khodesh ba khodesh!
```

```python
Z=hierarchy.linkage(dist_matrix,'complete')
```

```python
dendro=hierarchy.dendrogram(Z)
```
